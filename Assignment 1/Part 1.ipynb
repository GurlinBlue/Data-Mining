Exercise 2:Handling Missing Data

```python
Data = pd.read_csv("C:/Users/User/Downloads/melb_data.csv/melb_data.csv")
Data.head()
```

```python
Data_dropna = Data.dropna()
print("Dataframe after dropping rows with missing values :")
print(Data_dropna)
```

```python
mean_year = Data['YearBuilt'].mean()
Data_mean_fill = Data.fillna({'YearBuilt':mean_year})
print("Dataframe after filling missing 'YearBuilt' with mean:")
print(Data_mean_fill)
```

```python
median_BuildingArea = Data['BuildingArea'].median()
Data_median_fill = Data.fillna({'BuildingArea':median_BuildingArea})
print("Dataframe after filling missing 'BuildingArea' with median:")
print(Data_median_fill)
```

```python
missingvalues_cnt = Data.isnull().sum()
print(missingvalues_cnt)
```

```python
Data_specific_fill = Data.fillna({'CouncilArea':'Unknown'})
print("Dataframe after filling missing 'CouncilArea' with 'Unknown':")
print(Data_specific_fill)
```

```python
Data_ffill = Data.ffill()
print(Data_ffill)
```

```python
Data_bfill = Data.bfill()
print(Data_bfill)
```

Exercise 3:Data Transformation

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
Data['Car'] = scaler.fit_transform(Data[['Car']])
print(Data)
```

```python
from sklearn.preprocessing import StandardScaler
Data['Price'] = StandardScaler().fit_transform(Data[['Price']])
print(Data)
```

```python
Data = pd.DataFrame(Data)
original_columns = Data.columns
Data_dummies = pd.get_dummies(Data ,columns = ['Regionname'])
new_columns = Data_dummies.columns.difference(original_columns)
new_Data_columns = Data_dummies[new_columns]
print(new_Data_columns)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop = 'first', sparse_output = False)
encoded_array = encoder.fit_transform(Data[['Regionname']])
encoded_Data = pd.DataFrame(encoded_array, columns = encoder.get_feature_names_out(['Regionname']))
Data_encoded = pd.concat([Data,encoded_Data], axis = 1).drop(columns=['Regionname'])
print(Data_encoded)
```

```python
Data['Price_range']=pd.cut(Data['Price'],bins = 4,labels = ['Cheap','Medium','Expensive','Ridiculously expensive'],right = True)
new_binned_columns = Data[['Suburb','Price_range']]
print(new_binned_columns.head(20))
print(new_binned_columns.tail(20))
```
